# Implementating NN to solve XOR through Keras

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```
# the four different states of the XOR gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```
model = Sequential() # sets up an empty model using the Sequential API
model.add(Dense(16, input_dim=2, activation='relu')) # We set input_dim=2 because each of our 
                                                     # input samples is an array of length 2 
                                                     # 16 neurons in a hidden layer
model.add(Dense(1, activation='sigmoid')) # output layer with one neuron

model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['binary_accuracy']) 

model.fit(training_data, target_data, nb_epoch=1500, verbose=2)

print(model.predict(training_data).round())
```
